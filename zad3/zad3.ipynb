```python
import numpy as np
import pandas as pd
```

```python
# Zadanie 1
df = pd.read_csv('PAD_03_PD.csv', delimiter=';')
df
```

```python
len(df[df.isna().any(axis=1)])
```

```python
# Wiersze z kolumnami bez wartości stanowią bardzo małą populację (22/800). Wiersze te usuwam.
df = df.dropna().reset_index()
```

```python
df.Country.value_counts()
```

```python
# Zadanie 2
goods_columns = tuple(column for column in df.columns if 'owns' in column)
goods_columns
```

```python
df['owned_goods'] = df.loc[:, goods_columns].sum(axis=1)
df
```

```python
# Zadanie 3
np.round(df[df['gender'] == 'M'].owned_goods.mean(), decimals=2)
```

```python
np.round(df[df['gender'] == 'K'].owned_goods.mean(), decimals=2)
```

```python
# Zadanie 4
grouped = df.groupby(['Country'], as_index=False).agg({'owned_goods': lambda x: np.round(np.mean(x), decimals=2), 'Age': 'min'})
grouped = grouped.rename(columns={'owned_goods': 'owned_goods_mean', 'Age': 'Age_min'})
grouped
```
