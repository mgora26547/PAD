```python
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv('orders.csv')
orders.head()
```

```python
orders.info()
```

```python
orders.describe(include='all')
```

```python
orders.order_date = pd.to_datetime(orders.order_date, format="%Y/%m/%d")
orders
```

```python
orders.tshirt_category.value_counts()
```

```python
def normalize_names(x):
    x = x.replace('Wh Tshirt', 'White T-Shirt')
    x = x.replace('Bl Tshirt', 'Black T-Shirt')
    return x
orders.tshirt_category = orders.tshirt_category.apply(normalize_names)
orders.tshirt_category.value_counts()
```

```python
import re
```

```python
tshirt_type = re.compile('Hoodie|T-Shirt|Tennis Shirt')
tshirt_color = re.compile('Black|White')
tshirt_size = re.compile('M|F$')
orders['tshirt_type'] = orders.tshirt_category.apply(lambda x: tshirt_type.search(x)[0] if (tshirt_type.search(x)) else np.nan)
orders['tshirt_color'] = orders.tshirt_category.apply(lambda x: tshirt_color.search(x)[0] if (tshirt_color.search(x)) else np.nan)
orders['tshirt_size'] = orders.tshirt_category.apply(lambda x: tshirt_size.search(x)[0] if (tshirt_size.search(x)) else np.nan)
orders
```

```python
import datetime
```

```python
date_before = pd.to_datetime('2014-12-31')
date_after = pd.to_datetime('2016-8-2')
orders.loc[(orders['order_date'] > date_before) & (orders['order_date'] < date_after), :]
```

```python
customers = pd.read_csv('customers.csv')
customers.columns
```

```python
customers.shape
```

```python
customers = customers.rename(columns={'customerID': 'customer_id'})
customers.columns
```

```python
merged = pd.merge(orders, customers, on='customer_id', how='inner')
merged
```

```python
"""
Użyto metody merge, ponieważ pozwala ona na podanie wspólnej kolumny (parametr "on") jako klucza.
Parametr "how" został ustawiony na "inner" aby uniknąć powstawania niepełnych rekordów.
"""
```
